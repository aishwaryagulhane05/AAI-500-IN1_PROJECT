## Model Selection and Analysis 

### ARIMA
#### Why ARIMA?
    - ARIMA is designed specifically for forecasting when data is ordered by time, making it ideal for sales trend prediction.
        - In our dataset we have data ordered by date. Making it a suitable dataset for sales trend prediction using ARIMA
    - It captures auto-regression (AR) (using prior observations), integration (I) (to remove trend), and moving averages (MA) (to smooth random fluctuations), making it well-suited for weekly sales data.

#### ARIMA (Weekly sales forecast)
        - The best ARIMA model (p=2,d=1,q=2) captures your weekly sales trend well. 
        - An error of roughly 25% suggests the model is doing a good job (within good range), especially given the variability that often occurs in sales data. 

#### ARIMA (Daily sales forecast)
        - There is no combination of pdq which generates SMAPE < 30. So, Daily Predictions are not performing well in terms of accuracy.
        - Possible Reason: daily data is highly granular and noisy - and is not yielding better accuracy metrics (lower MAPE) as well.
#### NOTE:
- #### Prediction: Future Sales using historical sales patterns.
- #### ARIMA (AutoRegressive Integrated Moving Average) is a powerful statistical method for forecasting time‑series data.
- #### ARIMA is ideal when you have one variable over time (e.g., total sales) and want to forecast future points based on its own historical patterns

#### Limitation & Scope for Improvement
- The dataset contains only 1 year of sales data, providing a limited number of training points for the ARIMA model.
- Despite this limitation, the model is still able to reliably capture weekly sales patterns.
- Future Scope: As more data becomes available (multiple years), the forecasting accuracy is expected to improve, allowing the ARIMA model to better learn long‑term trends and seasonality. 

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


# Checking if the output directory exists
output_dir = "/home/Isha/USD/usd_project_aai_500/AAI-500-IN1_PROJECT/visuals/modeling_charts"
os.makedirs(output_dir, exist_ok=True)

#  Loading Processed Data
df = pd.read_csv('/home/Isha/USD/usd_project_aai_500/AAI-500-IN1_PROJECT/data/Processed/01_data_cleaning.csv')
df = df[df["TotalSales"]>0]

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df
```

## WEEKLY FORCAST : ARIMA

```python
sales_data = df
```

#### Data Preparation for ARIMA

```python
import itertools
import warnings
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd
import matplotlib.dates as mdates

warnings.filterwarnings("ignore")  # To skip warning messages

df_arima = sales_data.copy()
df_arima['Date'] = pd.to_datetime(df_arima['Date'], dayfirst=True, errors='coerce')
df_arima = df_arima.dropna(subset=['Date'])
df_arima.set_index('Date', inplace=True)

# resampling using the DateIndex
daily_sales = df_arima['TotalSales'].resample('D').sum()
daily_sales = daily_sales.fillna(0)

# Weekly sales
weekly_sales = daily_sales.resample('W').sum()

# Weekly DataFrame with Week Number
weekly_data = weekly_sales.to_frame(name='TotalSales')

# Plotting the data
weekly_data.plot(figsize=(10,4), title='Weekly Total Sales')
# Save the figure
output_file = os.path.join(output_dir, "Weekly_total_sales.png")
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
print(f"output saved at: {output_file}")

# Check Stationarity (ADF Test)
adf_result = adfuller(daily_sales.dropna())
print("-----------------------------")
print(f"ADF Statistic: {adf_result[0]:.4f}")
print(f"p-value: {adf_result[1]:.4f}")

weekly_sales.head()

```

#### Split Test data and Train data

```python
print(f"Date range: {weekly_sales.index.min()} -> {weekly_sales.index.max()}")

# Train Test Split for weekly data
train_w = weekly_sales.loc[:'2019-08-31']
test_w = weekly_sales.loc['2019-09-01':]
print("-----------------------------")
print(f"Train range: {train_w.index.min()} -> {train_w.index.max()} ({len(train_w)} points)")  
print(f"Test range : {test_w.index.min()} -> {test_w.index.max()} ({len(test_w)} points)")
```

#### Iterating over different p, d, q values to fit ARIMA 

```python
p_values = range(0, 3)
d_values = range(0, 3) 
q_values = range(0, 3)  
results = []
forecast_steps = len(test_w) + 12 # adding future 12 weeks
print("forecast_steps =",forecast_steps)
# Grid search over p, d, q # daily_sales
for p, d, q in itertools.product(p_values, d_values, q_values):
    try:
        #print("inloop")
        model = ARIMA(train_w, order=(p, d, q))
        model_fit = model.fit()
        #print("fit")
        forecast = model_fit.forecast(steps=forecast_steps)
        actual_values = test_w
        #print("forecast")
        #print("pdq:", p,d,q)
        predicted_values = forecast
        mae = np.mean(np.abs(actual_values - predicted_values))
        mape = np.mean(np.abs(actual_values - predicted_values) / np.abs(actual_values) * 100)
        smape = 100 * np.mean(2 * np.abs(actual_values - predicted_values) /(np.abs(actual_values) + np.abs(predicted_values)))
        results.append({'p': p, 'd': d, 'q': q, 'AIC': model_fit.aic, 'MAE': mae,'MAPE':mape, 'SMAPE': smape, 'Forecast': predicted_values})
        #print(results)
    except:
        continue

# Printing and plotting results having MAPE <= 30
for res in results:
    if res['MAPE'] <= 30:
        p, d, q = res['p'], res['d'], res['q']
        prediction = res['Forecast']

        # Ensure prediction has the test_w index # **Used AI Assistant to Generate this part of code  
        if not isinstance(prediction, pd.Series):
            prediction = pd.Series(prediction, index=test_w.index)

        comparison = pd.DataFrame({
            'Actual': test_w,
            'Predicted': prediction,
            'Difference': test_w - prediction
        })

        print("---------------------------")
        print(f"Model: p={p}, d={d}, q={q} | AIC={res['AIC']:.2f} , MAE={res['MAE']:.2f}, MAPE={res['MAPE']:.2f}")
        print("Top 5 Actual vs Predicted:\n", comparison.head())

        # Plotting after creating the figure
        fig, ax = plt.subplots(figsize=(12, 5))
        ax.plot(train_w, label='Train', color='blue')
        ax.plot(test_w, label='Test (Actual)', color='green')
        ax.plot(prediction, label=f'Forecast p={p}, d={d}, q={q}', color='red')
        ax.set(title=f'ARIMA Forecast vs Actual (p={p}, d={d}, q={q})',
               xlabel="Date",
               ylabel="Total Sales")
        ax.legend()
        ax.grid(True)

        # Date formatting
        import matplotlib.dates as mdates
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
        ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
        for label in ax.get_xticklabels():
            label.set_rotation(45)

        # Save
        output_file = os.path.join(output_dir, f"ARIMA_weekly_total_sales_prediction_p={p}_d={d}_q={q}.png")
        plt.savefig(output_file, dpi=300, bbox_inches='tight')
        plt.show()

        print(f" Output saved at: {output_file}")


print("---------------------------")
best_model = min(results, key=lambda x: x['MAPE'])
print("Best Model Parameters: (p,d,q)", best_model['p'], best_model['d'], best_model['q'])
print("AIC:", best_model['AIC']) 
# AIC (Akaike Information Criterion):A measure of model quality — balances fit vs complexity - Lower is better.
## Shows how well model fits th#### The best ARIMA model (2,1,2) captures your weekly sales trend well. 
#### An error of roughly 25% suggests the model is doing a good job (within good range), especially given the variability that often occurs in sales data. e data relative to the number of parameters used.
print("MAE:", best_model['MAE'])
# MAE (Mean Absolute Error):
## The average absolute difference between actual and predicted values.Lower MAE = closer prediction.
print("MAPE:", best_model['MAPE'])
# MAPE (Mean Absolute Percentage Error):
## Percentage error between actual and predicted. MAPE < 30  is Good
# Plot after creating the figure

print("---------------------------")
## A low MAPE (and low AIC) means the model captures the data well and predicts reliably.
```

## DAILY FORCAST : ARIMA
